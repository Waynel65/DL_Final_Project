Need to have files of the same name (same utterance) across all folders (L1 speaker, L2 speaker, and model output).

```python
import os
import time
import random
from IPython.display import Audio, display, clear_output
import ipywidgets as widgets
from scipy.io import wavfile
import numpy as np

# select number of samples that each participant will be tested on
Nsamples = 10

# Accentedness Test

# select data folder
audio_path = '/content/drive/MyDrive/Electrical Engineering courses/Deep Learning/Project/test_accent/'

accentedness_ratings = []

for i, folder in enumerate(['L1', 'L2', 'model']):
    folder_path = os.path.join(audio_path, folder)
    
    # randomize samples
    wav_files = [file for file in os.listdir(folder_path) if file.endswith('.wav')]
    selected_files = random.sample(wav_files, Nsamples)

    for j, file in enumerate(selected_files):
        print('Accentedness Test\n')
        print(f"Block {i+1}:")
        print(f"Trial {j+1}/{Nsamples}")

        # display audio interface
        audio_file = os.path.join(folder_path, file)
        display(Audio(audio_file))
        
        # wait for audio interface to appear before displaying response prompt
        time.sleep(2)

        # display response prompt
        rating = 0
        while not (rating >= 1 and rating <= 9):
            rating = input("Enter rating (1: no accent, 9: heavy accent):\n")
            try:
              rating = int(rating)
            except:
              rating = 0
              print("Wrong option selected. Please try again.")
              
        # append data
        accentedness_ratings.append((folder, file, rating))

        # clear printed output for next trial
        clear_output(wait=True)

# Save results
np.save('accentedness_ratings.npy', accentedness_ratings)
```

```python
# Voice Identity Test

# select data folder
audio_path = '/content/drive/MyDrive/Electrical Engineering courses/Deep Learning/Project/test_identity/'

voice_similarity_ratings = []
for folder in ['model']:
    folder_path = os.path.join(audio_path, folder)

    # randomize samples
    wav_files = [file for file in os.listdir(folder_path) if file.endswith('.wav')]
    selected_files = random.sample(wav_files, Nsamples)

    for k, file in enumerate(selected_files):
        
        print('\nVoice Identity Test\n')
        print(f"Trial {k}/{Nsamples} \n")

        if file.endswith('.wav'):

            # model output sample
            audio_file = os.path.join(folder_path, file)
            fs, data = wavfile.read(audio_file)
            data_reversed = data[::-1]
            wavfile.write('temp_reversed.wav', fs, data_reversed)
            
            # L1 speaker sample
            ref_audio_file = os.path.join(audio_path, 'L1', file)
            ref_fs, ref_data = wavfile.read(ref_audio_file)
            ref_data_reversed = ref_data[::-1]
            wavfile.write('temp_ref1_reversed.wav', ref_fs, ref_data_reversed)
            
            # L2 speaker sample
            ref_audio_file = os.path.join(audio_path, 'L2', file)
            ref_fs, ref_data = wavfile.read(ref_audio_file)
            ref_data_reversed = ref_data[::-1]
            wavfile.write('temp_ref2_reversed.wav', ref_fs, ref_data_reversed)
            
            # randomize order of L1 and L2 speakers
            if random.choice([True, False]):
                print("Speaker 1:")
                display(Audio('temp_ref1_reversed.wav'))
                print("Speaker 2:")
                display(Audio('temp_ref2_reversed.wav'))
                print("Test:")
                display(Audio('temp_reversed.wav'))
                order = 'L1-L2'
            else:
                print("Speaker 1:")
                display(Audio('temp_ref2_reversed.wav'))
                print("Speaker 2:")
                display(Audio('temp_ref1_reversed.wav'))
                print("Test:")
                display(Audio('temp_reversed.wav'))
                order = 'L2-L1'
                
            # save order of speakers
            order_options = order.split('-')

            # get decision about voice identity
            decision = widgets.Dropdown(options=['Speaker 1', 'Speaker 2'], description='Choose:')
            display(decision)
            
            time.sleep(3)

            # get confidence report
            confidence = 0
            while not (confidence >= 1 and confidence <= 7):
                confidence = input("Confidence (1:low, 7:high): \n")
                try:
                  confidence = int(confidence)
                except:
                  confidence = 0
                  print("Wrong option selected. Please try again.")

            # map voice identity report to order of speaker shown
            if decision.value == 'Speaker 1':
                decision_final = order_options[0]
            elif decision.value == 'Speaker 2':
                decision_final = order_options[1]

            # get vss score
            if decision_final == 'L1':
                vss = -confidence
            else:
                vss = confidence
            
            # append data
            voice_similarity_ratings.append((folder, file, vss, order))

            # clear printed output for next trials
            clear_output(wait=True)

# Save results
np.save('voice_similarity_ratings.npy', voice_similarity_ratings)

# Clean up temporary files
os.remove('temp_reversed.wav')
os.remove('temp_ref1_reversed.wav')
os.remove('temp_ref2_reversed.wav')
```
